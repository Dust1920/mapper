```python
new_name = 'Prueba'
new_df = template12

temp_path = 'templates\\26.xlsx'
def update_templates(temp_path, new_template):
    new_name = new_template[0]
    new_df = new_template[1]
    ofile = pd.read_excel(temp_path,sheet_name=None, index_col=0)
    oks = list(ofile.keys())
    ldf = [ofile[k] for k in oks]
    names = oks
    try:
        i_n = oks.index(new_name)
        ldf[i_n] = new_df
    except Exception as e:
        ldf.append(new_df)
        names.append(new_name)
        print(e)
    templates = pd.ExcelWriter(temp_path)
    for n, df in enumerate(ldf):
        df.to_excel(templates, sheet_name=names[n], index = 'NOMGEO')
    templates.close()

update_templates(temp_path, [new_name, new_df])
```
